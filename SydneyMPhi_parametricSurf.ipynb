```python
import numpy as np
import matplotlib.pyplot as plt
# generate 3d plot using mplot3d toolkit
# mpl_toolkits.mplot3d.axes3d.Axes3D.plot_surface(x, y, z)
import mpl_toolkits.mplot3d as Axes3D


# R = major radius = distance b/w the center of the torus & the center of the tube
R = 2
# r = minor radius = the radius of the tube
r = 1
# theta represents the rotation about the center of the torus (moving the orange circle)
# phi represents the rotation about the axis of revolution (moving the horizontal circle)
tht = phi = np.linspace(0, 2 * np.pi, 50)
# create a meshgrid of (theta, phi)
tht, phi = np.meshgrid(tht, phi)

# parametric eqn:
x = (R + r * np.cos(tht)) * np.cos(phi)
y = (r + r * np.cos(tht)) * np.sin(phi)
z = r * np.sin(tht)

# adjust the figure
fig = plt.figure('Parametric surfaces')
ax = fig.add_subplot(111, projection='3d')

# plot the surface
h = ax.plot_surface(x, y, z, cmap='jet', edgecolor='k')
# colorbar for easy visualization
fig.colorbar(h)
# add labels
ax.set_xlabel('x', fontweight='bold', fontsize=14)
ax.set_ylabel('y', fontweight='bold', fontsize=14)
ax.set_zlabel('z', fontweight='bold', fontsize=14)
ax.set_title('Parametric equation of a torus',
             fontweight='bold', fontsize=16)
ax.axis('square')

plt.show()
```

```python
# final code:
import numpy as np
import matplotlib.pyplot as plt


# parameters
# major radius = distance between the center of the torus and the center of the tube
major_radius = 4
# minor radius = the radius of the tube
minor_radius = 1
# axis limits:
upper_limit = major_radius + minor_radius
lower_limit = - upper_limit

# theta and phi represent the rotation about the center of the torus and the axis of revolution, respectively
# create a meshgrid of (theta, phi)
theta, phi = np.meshgrid(np.linspace(0, 2 * np.pi, 25), np.linspace(0, 2 * np.pi, 25))

# parametric equations
x = (major_radius + minor_radius * np.sin(phi)) * np.cos(theta)
y = (major_radius + minor_radius * np.sin(phi)) * np.sin(theta)
z = minor_radius * np.cos(phi)

# adjust the figure
fig = plt.figure('Parametric surfaces', figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# plot the surface
torus = ax.plot_surface(x, y, z,
                        cmap='jet', edgecolor='k')
# color bar for easy visualization
fig.colorbar(torus)
# axis labels
ax.set_xlabel('x', fontweight='bold', fontsize=14)
ax.set_ylabel('y', fontweight='bold', fontsize=14)
ax.set_zlabel('z', fontweight='bold', fontsize=14)
ax.set_title('Parametric equation of a torus', fontweight='bold', fontsize=16)
# set constrained scaling  to ensure that the torus looks right

ax.set_xlim(lower_limit, upper_limit)
ax.set_ylim(lower_limit, upper_limit)
ax.set_zlim(lower_limit, upper_limit)

# In PyCharm, matplotlib plots are displayed in the "SciView" tool window
# SciView does not currently support the interactive manipulation (like rotation) of 3D plots
# Change the settings to display plots in a separate window that supports interactive manipulation
# press Cmd + shift + A or Ctrl + Shift + A => search 'Python Scientific"
# uncheck the 'Show plots in tool window' option
plt.show()
```

```python
# ref code from wikipedia:
from mpl_toolkits.mplot3d import axes3d
from matplotlib.patches import Circle
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.art3d as art3d
import numpy as np


R, r = 15, 5


def get_x(phi, theta):
    global r, R
    return (R + r * np.cos(phi)) * np.cos(theta)

def get_y(phi, theta):
    global r, R
    return (R + r * np.cos(phi)) * np.sin(theta)

def get_z(phi, theta):
    global r
    return r * np.sin(phi)


if __name__ == "__main__":
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')

    ph = th = np.linspace (-np.pi, np.pi,30)
    phi, theta = np.meshgrid (ph, th)

    x = np.array([get_x(p,t) for p,t in zip(np.ravel(phi), np.ravel(theta))])
    X = x.reshape(phi.shape)
    y = np.array([get_y(p,t) for p,t in zip(np.ravel(phi), np.ravel(theta))])
    Y = y.reshape(phi.shape)
    z = np.array([get_z(p,t) for p,t in zip(np.ravel(phi), np.ravel(theta))])
    Z = z.reshape(phi.shape)
    C1 = Circle( (R,0), radius=r, fc='none', ec='red')
    C2 = Circle( (0,0), radius=R, fc='none', ec='magenta')

    ax.plot_wireframe(X, Y, Z, color='#CCCCCC')
    ax.add_patch(C1)
    ax.add_patch(C2)
    art3d.pathpatch_2d_to_3d(C1, z=0, zdir='y')
    art3d.pathpatch_2d_to_3d(C2, z=r, zdir='z')

    ax.set_xlim3d(-20,20)
    ax.set_ylim3d(-20,20)
    ax.set_zlim3d(-20,20)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_axis_off()

    ax.view_init(elev=30, azim=-45)
    plt.show()
    plt.savefig('torus.svg', bbox_inches='tight', pad_inches=.15, \
        bbox_extra_artists=[C1,C2], transparent='true')
```
